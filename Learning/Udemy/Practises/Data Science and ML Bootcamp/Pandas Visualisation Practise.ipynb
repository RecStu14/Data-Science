# Pandas Visualisation Practise

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#calling the df without the index_col argument
df1 = pd.read_csv("df1")
```

```python
df1.head()
#you would get an unnamed col and another extra column as the index col
#hence you should put the index_col argument
```

```python
df1 = pd.read_csv("df1",index_col=0)
```

```python
df1.head()
#this is a time series
```

```python
df2 = pd.read_csv("df2")
```

```python
df2.head() #Non-time series df
```

```python
#to change the style call seaborn
sb.set_style('darkgrid')
```

```python
#to get a histogram of all the values in the a column
df1["A"].hist() #actually calling matplotlib 
```

```python
df1["A"].hist(bins=30) #actually calling matplotlib 
```

```python
#another way to create hist (.plot)
df1['A'].plot(kind='hist')
```

```python
#another way .plot.hist() instead of kind
df1['A'].plot.hist()
```

```python
#area plot
df2.plot.area()
```

```python
#area plot - with transparency
df2.plot.area(alpha = 0.4)
```

```python
df2.plot.bar()
#the index value is taken as a categorical data
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line( y='B', figsize=(12,3),lw=1) #works if you leave out x
```

```python
df1.plot.scatter(x='A',y='B', c='C',cmap='coolwarm')
#showing a 3D plot 
```

```python
df1.plot.scatter(x='A',y='B', c='C',s=df1['C']*100)
#showing a 3D plot , the size indicates how large the C value 
#was relative to each other
```

```python
#Boxplots
df2.plot.box()
```

```python
#hexagonal bin plot
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.plot.hexbin(x='a',y='b')
```

```python
#enlarge the grid
df.plot.hexbin(x='a',y='b',gridsize=25)
```

```python
#KDE
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```

```python
#of all the columns in df2
df2.plot.kde()
```
