## Creating Plots

```python
#creating a plot
#a) functional method
#b) object oriented method
```

### Functional Method

```python
#functional method
plt.plot(x,y) #plotted x vs y
plt.show()
```

```python
#adding label and titles
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

### Object Oriented Method (Recommended)

```python
#Object Oriented Method (Better)
#Getting a blank canvas
fig = plt.figure()

#Adding Axes to the Canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #adding an axes, by passing a list containing the 
#value of left,bottom,width and height, %of the whitespace

#Plotting the graph
axes.plot(x,y)

#Adding labels
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

## Creating Multiplots

### Functional Method

```python
#multiplots on the same canvas
plt.subplot(1,2,1) #no of rows, no of cols, plot no u r referring to
plt.plot(x,y,'r') #r is colour

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object Oriented Method

```python
#Multiplot 

#Canvas
fig = plt.figure()

#Getting axes
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left,bottom, width, ht
#all these values should be between 0 and 1

#Plotting
axes1.plot(x,y)
axes2.plot(y,x)

#Labels
axes2.set_title("Smaller Plot")
axes1.set_title("Larger Plot")

#and then you manually adjust the axes to position the 2nd plot
```

### Subplots

```python
#The difference between a funcitonal/OOM and a subplot is that for the 
#functional/OOM you have to manually code in the figure and the axes but for
#subplots, the axes and figure already are automatically created!
```

```python
#creating subplots
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(1,2)
#in subplots you can specify the no of rows and cols 

#when stuff overlaps, to set it right
plt.tight_layout()
```

```python
axes #is just an array of the axes, and can be iterated thru
```

```python
#using a for loop to plot (iterating)
fig,axes=plt.subplots(1,2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
#plotting via indexing
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

## Figure Size, Aspect Ratio and DPI

```python
fig = plt.figure(figsize=(3,2)) #where width=3, ht=2, dpi is the pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2)) #where width=3, ht=2, dpi is the pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

#upon changing the figure size
```

```python
fig,axes = plt.subplots(figsize=(8,2)) #where width=3, ht=2, dpi is the pixels per inch

axes.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(2,1,figsize=(8,2)) #where width=3, ht=2, dpi is the pixels per inch

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Legends

```python
#Legends

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.set_title("Title")

ax.plot(x,x**3,label="X Cubed")

#adding a legend

#calling legend
ax.legend() #need to edit the plot call on top

#if legend overlaps the graph then 
ax.legend(loc=0), #0 is for the best location
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00',linewidth=3,alpha=0.5,linestyle='-.') #RGB Hex Code, the linewidth increases 
#the thickness, alpha controls the transparency, instead of writing linewidth
#you can type lw=,linestyle(can be replaced by ls) 
```

```python
#if you want to mark where the values occur
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='green',linewidth=1,alpha=1,linestyle='-',marker='o',markersize=10,
       markerfacecolor='blue',markeredgewidth=3,markeredgecolor='orange')
```

```python
#to show only a portion of the graph 
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='green',linewidth=2,alpha=0.5,linestyle='--')

#setting the x axis
ax.set_xlim([0,1]) #lower bound and upper bound for the x axis

#setting the y axis
ax.set_ylim([0,2]) #lower bound and upper bound for the y axis
```
